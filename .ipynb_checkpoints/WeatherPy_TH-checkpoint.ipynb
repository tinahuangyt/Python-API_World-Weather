# WeatherPy 

* As expected, the maximum temperature are observed in cities at or near the equator (latitude equals or close to 0). As the cities get farther away from the equator (latitude approaching + or - 90), the maximal temperature drops. Unsurprisingly, the northern hemisphere is currently much colder than the southern hemisphere. 

* Currently, cities in the northern hemisphere have a wide and even spread of humidity, whereas in the southern hemisphere, more than 2/3 of cities have humidity above 80%. 

* There is no observable correlation between latitude and cloudiness and between latitude and windspeed. There is an even distribution of cloudless cities across a range of latitude (-50 to 80). Most cities currently have less than 15mph of windspeed. 


This analysis is completed with full awareness that there are much fewer cities falling below latitude of -60, and can lead to uneven sampling. Conversly, there are a lot more cities in the northern hemisphere between +30 to +60 latitude. Data should be interpreted with this in mind.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API calls
* Weather data obtained from OpenWeatherMap API on March 18, 2019.

```python
#Set up API call parameters

url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

query_url = f"{url}appid={api_key}&units={units}&q="
city = cities

#Each set will have 50 entries
x=1
y=1

city_name =[]
latitude = []
maxtemp = []
humidity = []
cloudiness = []
windspeed = []

for city in cities:
    response = requests.get(query_url + city).json()
    try:
        city_name.append(response["name"])
        latitude.append(response["coord"]["lat"])
        maxtemp.append(response["main"]["temp_max"])
        humidity.append(response["main"]["humidity"])
        cloudiness.append(response["clouds"]["all"])
        windspeed.append(response["wind"]["speed"])
        print(f"Processing record number {x} of set {y} | {city}")
        if x>49:
            x=1
            y=y+1
        else:
            x=x+1
    except:
        print(f"City not found, next..")
```

```python
#Combine all lists to generate CSV file
combined_list = list(zip(city_name, latitude, maxtemp, humidity, cloudiness, windspeed))

#Export to CSV
import csv
myfile = open("world_weather_info.csv",'w')
with myfile:
    header = ["City", "Latitude", "Max_Temp (F)","Humidity", "Cloudiness", "Windspeed (mph)"]
    writer= csv.writer(myfile)
    writer.writerow(header)
    writer.writerows(combined_list)
```

```python
#Create a df for all the retrieved data
world_weather = pd.DataFrame(columns=["City", "Latitude", "Max_Temp (F)","Humidity", "Cloudiness", "Windspeed (mph)"])
world_weather["City"]=city_name
world_weather["Latitude"]=latitude
world_weather["Max_Temp (F)"]=maxtemp
world_weather["Humidity"]=humidity
world_weather["Cloudiness"]=cloudiness
world_weather["Windspeed (mph)"]=windspeed

world_weather.head()
```

```python
#Count entries to ensure n>500
world_weather.count()
```

## Plotting data

#### Latitude vs. Max Temperature (F) Plot

```python
# Max temperature vs. Latitude

plt.scatter(world_weather["Latitude"], world_weather["Max_Temp (F)"],edgecolors = 'black', linewidths=1, alpha=0.75)

plt.title("City Latitude vs Max Temperature (3/18/19)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.grid()

plt.savefig("lat_maxtemp.png")
plt.show()
```

#### Latitude vs. Humidity (%) Plot

```python
# Humidity vs. Latitude

plt.scatter(world_weather["Latitude"], world_weather["Humidity"],edgecolors = 'black', linewidths=1, alpha=0.75)

plt.title("City Latitude vs Humidity (3/18/19)")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.ylim(0, 110, 20)
plt.grid()

plt.savefig("lat_humidity.png")
plt.show()
```

#### Latitude vs. Cloudiness (%) Plot

```python
# Cloudiness vs. Latitude

plt.scatter(world_weather["Latitude"], world_weather["Cloudiness"],edgecolors = 'black', linewidths=1, alpha=0.75)

plt.title("City Latitude vs Cloudiness (3/18/19)")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.grid()

plt.savefig("lat_cloudiness.png")
plt.show()
```

#### Latitude vs. Wind Speed (mph) Plot

```python
# Wind Speed vs. Latitude

plt.scatter(world_weather["Latitude"], world_weather["Windspeed (mph)"],edgecolors = 'black', linewidths=1, alpha=0.75)

plt.title("City Latitude vs Windspeed (3/18/19)")
plt.xlabel("Latitude")
plt.ylabel("Windspeed (mph)")
plt.grid()

plt.savefig("lat_windspeed.png")
plt.show()
```
